```python
# ensure you have the HF_TOKEN value(env) set to be able to run this
from transformers import pipeline
import torch

pipe = pipeline(
    "text-generation",
    model="meta-llama/Llama-3.2-1B-Instruct",
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
# uncomment the below messages to test one by one

messages = [
    {"role": "system", "content": "You are a text classifier and will classify the given text input as Lead or Not a Lead. Do not provide any explanation just respond with a Yes or a No."},
    {"role": "user", "content": "I love this product. Can i know more about it"},
    #{"role": "user", "content": "I dont think i want any information."},
    #{"role": "user", "content": "Ya you really think this solve my problem. I doubt it."},
]
outputs = pipe(messages, max_new_tokens=256)
print(outputs[0]["generated_text"][-1])
```
